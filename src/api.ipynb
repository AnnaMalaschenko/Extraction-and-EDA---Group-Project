```python
import os
import requests
import pandas as pd
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv("API_KEY")
viewName = 'cats'
limit = 250
page = 1

url = (
    f"https://api.rescuegroups.org/v5/"
    f"public/animals/"
    f"search/{viewName}/"
    f"?limit={limit}"
    f"&page={page}"
)

headers = {
    "Content-Type": "application/vnd.api+json",
    "Authorization": API_KEY
}
```

```python
response = requests.get(url, headers=headers)
```

```python
response.status_code
```

```python
data = response.json()
data
```

```python
data.keys()
```

```python
animal_data = data["data"]
type(animal_data)
```

```python
# animal_data is a list of dictionaries. Each dictionary is an individual animal. Each dictionary has the keys: type, id and attributes. Attributes is a key to another dictionary.

animal_data
```

```python
# I asked Claude to confirm my above understanding is correct. It explained how to convert this JSON into rows:

records = [] # create an empty list that will hold the rows I'm creating

for animal in animal_data:
    row = {"id": animal["id"]} # take the id and make it the start of the row
    row.update(animal["attributes"]) # take the attribute values and add them to the row
    records.append(row) # add each row created to the growing list of animals

# records is now a non-empty list, with each individual animal being represented by a dictionary inside the list. Here is what the printed list looks like:
for row in records:
    print(row)
```
